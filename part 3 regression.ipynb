## The Models:

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
# Bild a new DataFrame without the id coulomn
cars = pd.DataFrame(car.iloc[:,1:12])
cars
```

```python
X = cars.drop('current price',axis=1)
y = cars['current price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

###  Linear Regression

```python
linearR = LinearRegression()
linearR.fit(X_train,y_train)
pred_linear = linearR.predict(X_test)
```

```python
r2 = r2_score(pred_linear,y_test)
MAE = mean_absolute_error(pred_linear,y_test)
MSE = mean_squared_error(pred_linear,y_test)
RMSE = np.sqrt(MSE)
```

```python
print('r2: ' + str(r2))
print('MAE: ' + str(MAE))
print('MSE: ' +str(MSE))
print('RMSE: ' + str(RMSE))
```

```python
plt.scatter(pred_linear,y_test)
plt.xlabel('\n The data from the model')
plt.ylabel('The test data \n')
```

### KNN

```python
accuracy = []
for i in range (1,50):
    model_knn = KNeighborsRegressor(n_neighbors = i)
    model_knn_fit = model_knn.fit(X_train, y_train)
    model_knn_pred = model_knn.predict(X_test)
    acc_knn = r2_score(model_knn_pred, y_test)
    accuracy.append(round(acc_knn,3)*100)
index = accuracy.index(max(accuracy))+1 
print("The best r2 score with " + str(index) +" nearest neighbers")
plt.plot(range(1,50),accuracy)
plt.show()
```

```python
knn = KNeighborsRegressor(n_neighbors=4)

knn.fit(X_train,y_train)
pred = knn.predict(X_test)

r2 = r2_score(pred,y_test)
MAE = mean_absolute_error(pred,y_test)
MSE = mean_absolute_error(pred,y_test)
RMSE = np.sqrt(mean_absolute_error(pred,y_test))
```

```python
print('WITH K=4')
print('r2: ' + str(r2))
print('MAE: ' + str(MAE))
print('MSE: ' +str(MSE))
print('RMSE: ' + str(RMSE))
```

```python
plt.scatter(pred,y_test)
plt.xlabel('\n The data from the model')
plt.ylabel('The test data \n')
```
